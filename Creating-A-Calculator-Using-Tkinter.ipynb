```python
# Creating a Calculator using Tkinter

from tkinter import *
def click(event):
    global scvalue
    text=event.widget.cget("text")  # To get the text value of the button clicked on
    if text == "=":
        if scvalue.get().isdigit():
            value=int(scvalue.get())
        else:
            value=eval(screen.get())
        scvalue.set(value)
        screen.update()
    elif text =="AC":
        scvalue.set("")
        screen.update()
    else:
        scvalue.set(scvalue.get() + text)
        screen.update()

    
root=Tk()
root.geometry("400x550")
root.title("Calculator")
scvalue=StringVar()
scvalue.set("")

screen=Entry(root,textvar=scvalue,font="Lucida 30 bold",fg="white",bg="black")
screen.pack(fill='x',ipadx=8,pady=10,padx=10)
f0=Frame(root,bg='white')
f1=Frame(root,bg='white')
f2=Frame(root,bg='white')
f3=Frame(root,bg="white")
f4=Frame(root,bg="white")

b=Button(f0,text='(',font ="Lucida 30 bold",bg='black',fg='white',width=2)
b.pack(side=LEFT,padx=6,pady=6)
b.bind("<Button-1>",click)
b=Button(f0,text=')',font ="Lucida 30 bold",bg='black',fg='white',width=2)
b.pack(side=LEFT,padx=6,pady=6)
b.bind("<Button-1>",click)
b=Button(f0,text='%',font ="Lucida 30 bold",bg='black',fg='white',width=2)
b.pack(side=LEFT,padx=6,pady=6)
b.bind("<Button-1>",click)
b=Button(f0,text='AC',font ="Lucida 30 bold",bg='black',fg='white',width=2)
b.pack(side=LEFT,padx=6,pady=6)
b.bind("<Button-1>",click)
f0.pack()

b=Button(f1,text='7',font ="Lucida 30 bold",bg='grey',fg='white',width=2)
b.pack(side=LEFT,padx=6,pady=6)
b.bind("<Button-1>",click)
b=Button(f1,text='8',font ="Lucida 30 bold",bg='grey',fg='white',width=2)
b.pack(side=LEFT,padx=6,pady=6)
b.bind("<Button-1>",click)
b=Button(f1,text='9',font ="Lucida 30 bold",bg='grey',fg='white',width=2)
b.pack(side=LEFT,padx=6,pady=6)
b.bind("<Button-1>",click)
b=Button(f1,text='/',font ="Lucida 30 bold",bg='black',fg='white',width=2)
b.pack(side=LEFT,padx=6,pady=6)
b.bind("<Button-1>",click)
f1.pack()

b=Button(f2,text='6',font ="Lucida 30 bold",bg='grey',fg='white',width=2)
b.pack(side=LEFT,padx=6,pady=6)
b.bind("<Button-1>",click)
b=Button(f2,text='5',font ="Lucida 30 bold",bg='grey',fg='white',width=2)
b.pack(side=LEFT,padx=6,pady=6)
b.bind("<Button-1>",click)
b=Button(f2,text='4',font ="Lucida 30 bold",bg='grey',fg='white',width=2)
b.pack(side=LEFT,padx=6,pady=6)
b.bind("<Button-1>",click)
b=Button(f2,text='*',font ="Lucida 30 bold",bg='black',fg='white',width=2)
b.pack(side=LEFT,padx=6,pady=6)
b.bind("<Button-1>",click)
f2.pack()

b=Button(f3,text='1',font ="Lucida 30 bold",bg='grey',fg='white',width=2)
b.pack(side=LEFT,padx=6,pady=6)
b.bind("<Button-1>",click)
b=Button(f3,text='2',font ="Lucida 30 bold",bg='grey',fg='white',width=2)
b.pack(side=LEFT,padx=6,pady=6)
b.bind("<Button-1>",click)
b=Button(f3,text='3',font ="Lucida 30 bold",bg='grey',fg='white',width=2)
b.pack(side=LEFT,padx=6,pady=6)
b.bind("<Button-1>",click)
b=Button(f3,text='-',font ="Lucida 30 bold",bg='black',fg='white',width=2)
b.pack(side=LEFT,padx=6,pady=6)
b.bind("<Button-1>",click)
f3.pack()

b=Button(f4,text='0',font ="Lucida 30 bold",bg='grey',fg='white',width=2)
b.pack(side=LEFT,padx=6,pady=6)
b.bind("<Button-1>",click)
b=Button(f4,text='.',font ="Lucida 30 bold",bg='grey',fg='white',width=2)
b.pack(side=LEFT,padx=6,pady=6)
b.bind("<Button-1>",click)
b=Button(f4,text='=',font ="Lucida 30 bold",bg='darkblue',fg='white',width=2)
b.pack(side=LEFT,padx=6,pady=6)
b.bind("<Button-1>",click)
b=Button(f4,text='+',font ="Lucida 30 bold",bg='black',fg='white',width=2)
b.pack(side=LEFT,padx=6,pady=6)
b.bind("<Button-1>",click)
f4.pack()

root.mainloop()
```
