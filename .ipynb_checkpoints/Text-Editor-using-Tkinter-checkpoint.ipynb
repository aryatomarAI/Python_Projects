```python
# Creating a Text Editor Using Tkinter
from tkinter import *
from tkinter.filedialog import askopenfilename
from tkinter.filedialog import asksaveasfilename
import tkinter.messagebox as tmsg
import os
def new_file():
    global file
    root.title("Untitled - NoteWithUs")
    file=None
    text_area.delete(1.0,END)
    

def open_file():
    global file
    file=askopenfilename(defaultextension=".txt",filetypes=[("All files","*.*"),("Text Documents","*.txt")])
    if file=="":
        file=None
    else:
        root.title(os.path.basename(file) + " - NoteWithUs")
        text_area.delete(1.0,END)
        f=open(file,"r")
        text_area.insert(1.0,f.read())
        f.close()
    

def save_file():
    global file
    
    if file==None:
        file=asksaveasfilename(initialfile="Untitled.txt", defaultextension=".txt",filetypes=[("All files","*.*"),("Text Documents","*.txt")])
        if file=="":
            file=None
        else:
            f=open(file,"w")
            f.write(text_area.get(1.0,END))
            f.close()
            root.title(os.path.basename(file) + " - NoteWithUs")
    else:
        # Save the file
        with open(file,"w") as f:
            f.write(text_area.get(1.0,END))
            f.close()
            

def quit_app():
    root.destroy();

def cut_func():
    text_area.event_generate(("<<Cut>>"))

def copy_func():
    text_area.event_generate(("<<Copy>>"))

def paste_func():
    text_area.event_generate(("<<Paste>>"))

def undo_func():
    pass

def change_font():
    pass

def help_func():
    tmsg.showinfo("Need Help","Please visit our website for help!!!")

def about_func():
    tmsg.showinfo("About Us","This is a demo text editor which i have built with the help of tkinter GUI")

def send_feed():
    value=tmsg.askquestion("Rate Us","Was Your Expirence Good?")
    if value=='yes':
        msg="Great, rate us on appstore please !!"
    else:
        msg="Tell us what went wrong. We will contact you soon"
    tmsg.showinfo("NoteWithUs", msg)
    


if __name__=="__main__":
    root=Tk()
    root.geometry("500x600")
    root.title("Untitled - NoteWithUs")
    
    # Create a TextArea
    text_area=Text(root,font="lucida 13")
    file=None
    text_area.pack(expand=True,fill=BOTH)
    
    # Let's Create MenuBar
    menu_bar=Menu(root)
    # File Menu
    file_menu=Menu(menu_bar,tearoff=0)
    file_menu.add_command(label="New",command=new_file)
    file_menu.add_command(label="Open",command=open_file)
    file_menu.add_command(label="Save",command=save_file)
    file_menu.add_separator()
    file_menu.add_command(label="Exit",command=quit_app)
    menu_bar.add_cascade(label="File",menu=file_menu)
    
    # Edit Menu
    edit_menu=Menu(menu_bar,tearoff=0)
    edit_menu.add_command(label="Cut",command=cut_func)
    edit_menu.add_command(label="Copy",command=copy_func)
    edit_menu.add_command(label="Paste",command=paste_func)
    edit_menu.add_command(label="Undo",command=undo_func)
    menu_bar.add_cascade(label="Edit",menu=edit_menu)
    
    # Format Menu
    font_menu=Menu(menu_bar,tearoff=0)
    font_menu.add_command(label="Font..",command=change_font)
    menu_bar.add_cascade(label="Format",menu=font_menu)
    
    # Help menu
    help_menu=Menu(menu_bar,tearoff=0)
    help_menu.add_command(label="View Help",command=help_func)
    help_menu.add_command(label="Send Feedback",command=send_feed)
    help_menu.add_separator()
    help_menu.add_command(label="About NoteWithUs",command=about_func)
    menu_bar.add_cascade(label="Help",menu=help_menu)
    
    root.config(menu=menu_bar)
    
    scroll=Scrollbar(text_area)
    scroll.pack(side=RIGHT,fill='y')
    scroll.config(command=text_area.yview)
    text_area.config(yscrollcommand=scroll.set)
    
    root.mainloop()
    
```
